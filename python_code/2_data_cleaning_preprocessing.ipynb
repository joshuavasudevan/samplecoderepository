## Merging data

```python
#Import relevant libraries

import pandas as pd
import os
```

```python
# path to the folder containing the csv files
csv_dir = 'D:\\OneDrive - Loughborough University\\PhD\\Case study\\Crest\\data_ai_model\\crest_office_data_aq1\\aq1'
```

Now initialize an empty list to store dataframes. Next, loop through each file in the directory using the os.listdir() function. For each file, it checks if the file ends with ".csv" to make sure it is a CSV file. If it is, it loads the file into a dataframe using the pd.read_csv() function and appends the dataframe to the list. Finally, it concatenates all dataframes in the list into a single dataframe using the pd.concat() function. The resulting concatenated dataframe is stored in the variable aq1. Similarly aq2, aq3, aq4 are created.

```python
# Initialize an empty list to store dataframes
empty_list = []

# Loop through each file in the directory
for file in os.listdir(csv_dir):
    # Check if the file is a CSV file
    if file.endswith(".csv"):
        # Load the CSV file into a dataframe
        df = pd.read_csv(os.path.join(csv_dir, file))
        # Append the dataframe to the list
        empty_list.append(df)

# Concatenate all dataframes in the list into a single dataframe
aq1 = pd.concat(empty_list, ignore_index=True)
```

## Cleaning data 

Next,loop through the list of data frames using a for loop, and for each data frame, we call the drop() function to drop the specified columns

```python
# Dataframes are added as a list
aq_list = [aq1, aq2, aq3, aq4]

# Loop through the list of data frames and drop the specified columns
for df in aq_list:
    df.drop(['device_id', 'device_label', 'count', 'humidity', 'pm01', 'pm25', 'pm04', 'pm10'], axis=1, inplace=True)
```

```python
aq1 
```

Rename the columns to include the sensor names

```python
# rename the columns
aq1 = aq1.rename(columns={'normalized_reading_date': 'timestamp', 'co2': 'co2_aq1','temperature': 'temp_aq1', 'tvoc': 'tvoc_aq1'})
aq2 = aq2.rename(columns={'normalized_reading_date': 'timestamp', 'co2': 'co2_aq2','temperature': 'temp_aq2', 'tvoc': 'tvoc_aq2'})
aq3 = aq3.rename(columns={'normalized_reading_date': 'timestamp', 'co2': 'co2_aq3','temperature': 'temp_aq3', 'tvoc': 'tvoc_aq3'})
aq4 = aq4.rename(columns={'normalized_reading_date': 'timestamp', 'co2': 'co2_aq4','temperature': 'temp_aq4', 'tvoc': 'tvoc_aq4'})
```

Change the timestamp to datetime format

```python
# loop through each data frame and convert the Timestamp column to DateTime format
dfs = [aq1, aq2, aq3, aq4]
for df in dfs:
    df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
aq1
```

Merge all the 4 aq sensors into a single dataframe

```python
# merge the four data frames using the timestamp column
from functools import reduce
dfs = [aq1, aq2, aq3, aq4]
df = reduce(lambda left,right: pd.merge(left,right,on='timestamp', how='inner'), dfs)
```

## Feature Engineering

Pearson correlation analysis is done to check the correlation between the variables

```python
import matplotlib.pyplot as plt
import seaborn as sb
pearsoncorr = df.corr(method='pearson')
pearsoncorr
```

```python
plt.figure(figsize = (16,5))
ax =sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

Feature engineer the timeseries element into the cross sectional data. We extract the date, time element and we engineer it into the data

```python
df['timestamp'] = pd.to_datetime(df.timestamp, format='%Y-%m-%d %H:%M:%S')
df["IsWeekday"] = df["timestamp"].dt.weekday < 5 
df.IsWeekday = df.IsWeekday.replace({True: 1, False: 0})
df["hourofday"] = df["timestamp"].dt.hour
from datetime import datetime
df["dayofyear"] = df["timestamp"].dt.dayofyear
df["dayofweek"] = df["timestamp"].dt.dayofweek
```
